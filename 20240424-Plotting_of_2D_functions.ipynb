## Plotting functions of 2 variables

```python
%matplotlib widget
```

```python
from sympy import *
```

```python
x,y = symbols('x,y')
f = exp(-(x**2 + y**2))
```

```python
plotting.plot3d(f,(x,-2,2),(y,-2,2))
```

```python
plotting.plot_contour(f,(x,-2,2),(y,-2,2))
```

## Example: Plot of coordinate function in R^2

```python
x1,x2 = symbols('x1,x2')
f = x1
```

```python
plotting.plot_contour(f,(x1,-2,2),(x2,-2,2))
```

```python
plotting.plot3d(f,(x1,-2,2),(x2,-2,2))
```

## Visualizing the function from page 17

```python
f = x1*sin(x2)
plotting.plot3d(f)
```

## Partial derivatives in Python

```python
from sympy import *
x,y = symbols('x,y')
```

```python
f = x**2 * y + 2*y
f
```

```python
# Partial derivative with respect to x
diff(f,x)
```

```python
# Partial derivative with respect to y
diff(f,y)
```
