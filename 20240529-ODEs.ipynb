## Solving the logistic differential equation with SymPy

```python
# Define p as SymPy function (which we do not yet know)
p = Function('p')
```

```python
# assuming k=1, pmax=1
t = symbols('t')
C1 = symbols('C1')
result = dsolve(diff(p(t),t) - p(t)*(1-p(t)),p(t))
result
```

```python
type(result)
```

```python
print(result)
```

```python
result.rhs
```

```python
plotting.plot(result.rhs.subs({C1:1}))
```

## Example on page 25 (discharging a capacitor)

```python
U = Function('U')
t, R, C, C1 = symbols('t,R,C,C1')
result = dsolve(diff(U(t),t) + 1/(R*C)*U(t), U(t))
result
```

```python
plotting.plot(result.rhs.subs({
    C1:1,
    C:1,
    R:1
}));
```
