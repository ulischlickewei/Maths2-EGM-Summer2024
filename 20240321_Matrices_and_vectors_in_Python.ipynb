# Vectors in Sympy

```python
from sympy import *
```

```python
v = Matrix([1,2,3,4])
display(v)
```

```python
# Sum of two vectors
w = Matrix([0,1,0,1])
v+w
```

```python
# Multiplication with a scalar
2*v
```

```python
# Dot product / scalar product
v.dot(w)
```

```python
display(v)
```

```python
display(w)
```

```python
# Norm
v.norm()
```

```python
sqrt(v.dot(v))
```

```python
# Angle between two vectors
acos(v.dot(w)/(v.norm()*w.norm()))
```

```python
round(acos(v.dot(w)/(v.norm()*w.norm())),2)
```

# Matrices in SymPy

```python
# Defining a matrix by passing a list of lists,
# corresponding to the rows of the matrix
A = Matrix([[2,3,5,6],[1,-2,4,5]])
display(A)
```

```python
# Defining a row vector
Matrix([[1,2,3,4]])
```

```python
# Transposing a matrix
A.T
```

```python
# Matrix multiplication
display(A)
```

```python
B = A.T
A*B
```

```python
B*A
```
