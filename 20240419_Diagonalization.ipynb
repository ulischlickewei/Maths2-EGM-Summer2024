## Relationship between quadratic forms and diagonalization

```python
P = Matrix([[1,2],
           [-2,1]])
P
```

```python
A = Matrix([[3,0],
            [0,7]])
A
```

```python
M = P*A*P.inv()
M
```

```python
from sympy.plotting import plot3d
x1,x2 = symbols('x1, x2')
plot3d(31/5*x1**2 + 16/5*x1*x2 + 19/5*x2**2, (x1,-2,2), (x2,-2,2))
```

```python
plot_implicit(Eq(31/5*x1**2 + 16/5*x1*x2 + 19/5*x2**2,1), (x1,-0.8,0.8), (x2,-0.8,0.8))
```

```python
P = M.diagonalize()[0]
P
```

```python
y1, y2 = symbols('y1, y2')
plot_implicit(Eq(3*y1**2 + 7*y2**2, 1),
 (y1,-0.8,0.8), (y2,-0.8,0.8))
```
