## Example on page 54

```python
from sympy import *
x,y = symbols('x,y')
```

```python
f = exp(-x**2/2)*cosh(y)
f
```

```python
grad = Matrix([diff(f,v) for v in [x,y]])
grad
```

```python
solve(grad,[x,y])
```

```python
nsolve(grad,[x,y],[1,1])
```

```python
plot(cosh(x),(x,-3,3))
```

```python
plot(sinh(x))
```

```python
plot(sinh(x),(x,-4,4))
```

```python
hessian(f,(x,y)).subs({x:0,y:0}).eigenvals()
```

## Exercise sheet 5, exercise 5

```python
a,b = symbols('a,b')
```

```python
Err = ((0 - (a+b*(-1)))**2 +
     (1 - (a+b*0))**2 +
     (2 - (a+b*1))**2 +
     (4 - (a+b*2))**2
)
Err
```

```python
grad = Matrix([diff(Err,v) for v in [a,b]])
grad
```

```python
solve(grad,(a,b))
```

```python
hessian(Err,(a,b)).eigenvals()
```
